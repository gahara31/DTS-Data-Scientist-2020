## Background
<p>Text documents are one of the richest sources of data for businesses.

Weâ€™ll use a public dataset from the BBC comprised of 2225 articles, each labeled under one of 5 categories: business, entertainment, politics, sport or tech.

The dataset is broken into 1490 records for training and 735 for testing. The goal will be to build a system that can accurately classify previously unseen news articles into the right category.</p>

## Import Library

```python
#import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Collection

```python
#import dataset
df_train = pd.read_csv('BBC News Train.csv')
```

## Data Preparation

```python
def LabelCategory (row):
   if row['Category'] == 'business' :
      return 0
   if row['Category'] == 'entertainment' :
      return 1
   if row['Category'] == 'politics':
      return 2
   if row['Category']  == 'sport':
      return 3
   if row['Category'] == 'tech':
      return 4
```

```python
#df.apply (lambda row: LabelCategory(row), axis=1)
```

```python
df_train['LabelCategory'] = df_train.apply (lambda row: LabelCategory(row), axis=1)
```

```python
#replace each category name to numeric value

#df_train['Category'].replace(to_replace='business', value=0, inplace=True)
#df_train['Category'].replace(to_replace='entertainment', value=1, inplace=True)
#df_train['Category'].replace(to_replace='politics', value=2, inplace=True)
#df_train['Category'].replace(to_replace='sport', value=3, inplace=True)
#df_train['Category'].replace(to_replace='tech', value=4, inplace=True)
```

```python
df_train.head()
```

```python
#Split data training & test (70% training, 30% test)
from sklearn.model_selection import train_test_split

X = df_train['Text']
Y = df_train['LabelCategory']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
print('Total rows count: ', len(df_train))
print('Total Data Train: ', len(X_train))
print('Total Data Test: ', len(X_test))
```

<b><u>CountVectorizer</u></b><br>
CountVectorizer yang disediakan oleh scikit-learn library digunakan untuk membuat vektor kalimat. Library ini akan mengambil kata-kata dari setiap kalimat dan menciptakan vocabulary dari semua kata unik dalam kalimat. Vocabulary ini kemudian dapat digunakan untuk membuat feature vector dari jumlah kemunculan setiap kata

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(strip_accents='ascii', token_pattern=u'(?ui)\\b\\w*[a-z]+\\w*\\b', lowercase=True, stop_words='english')
X_train_cv = cv.fit_transform(X_train)
X_test_cv = cv.transform(X_test)
```

```python
#print(X_train_cv)
#print(cv.get_feature_names())
#print(cv.vocabulary_)
```

<center><img src="cv_example.JPG"></center>

```python
#Example
text = df_train.loc[:1]['Text']
#text_len = df_train.loc[:1]['Text'].str.len().sum()

# Create a Vectorizer Object 
vectorizer = CountVectorizer(stop_words='english') 
  
vectorizer.fit(text) 
  
# Printing the identified Unique words along with their indices 
#print("Vocabulary: ", vectorizer.vocabulary_) 
print("Length of Vocabulary: ", len(vectorizer.vocabulary_)) 
  
# Encode the Document 
vector = vectorizer.transform(text) 
  
# Summarizing the Encoded Texts 
#print("Encoded Document is:") 
#print(vector.toarray())

df_cv = pd.DataFrame(vector.toarray(),columns=vectorizer.get_feature_names())
#cv_dataframe = pd.set_option('display.max_columns', None)
print('Data Shape : ', df_cv.shape)
df_cv
```

```python
#filter based on category
df_sport = df_train[df_train['Category'].str.contains("sport")]
df_business = df_train[df_train['Category'].str.contains("business")]
df_politics = df_train[df_train['Category'].str.contains("politics")]
df_entertainment = df_train[df_train['Category'].str.contains("entertainment")]
df_tech = df_train[df_train['Category'].str.contains("tech")]

#export to txt file
with open('sportnews.txt', 'w') as f:
    f.write(df_sport['Text'].str.cat(sep='\n'))

with open('businessnews.txt', 'w') as f:
    f.write(df_business['Text'].str.cat(sep='\n'))
    
with open('politicsnews.txt', 'w') as f:
    f.write(df_politics['Text'].str.cat(sep='\n'))

with open('entertainmentnews.txt', 'w') as f:
    f.write(df_entertainment['Text'].str.cat(sep='\n'))
    
with open('technews.txt', 'w') as f:
    f.write(df_tech['Text'].str.cat(sep='\n'))
    
#make stopwords
stopwords = set(STOPWORDS)
# add words to stopwords, because it's not relevant and it stated alot
stopwords.add('said') 
stopwords.add('will')

#cloudword setting
bbc_news = WordCloud(
    background_color='white',
    stopwords=stopwords
)

#import news file
sportclouds = open('sportnews.txt','r').read()
businessclouds = open('businessnews.txt','r').read()
politicsclouds = open('politicsnews.txt','r').read()
entertainmentclouds = open('entertainmentnews.txt','r').read()
techclouds = open('technews.txt','r').read()
```

## Modeling

```python
# fitting naive bayes to the training set
# multinomial Naive Bayes
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train_cv, Y_train)
Y_pred = model.predict(X_test_cv)
```

```python
#accuray score
from sklearn.metrics import accuracy_score 

print ("Accuracy : ", accuracy_score(Y_test, Y_pred))
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_pred)

f, ax = plt.subplots(figsize=(8,5))
sns.heatmap(cm, annot=True, fmt=".0f", ax=ax)
plt.xlabel("Predicited Label")
plt.ylabel("Actual Label")
plt.show()
```

```python
#classification report
from sklearn.metrics import classification_report

print(classification_report(Y_test, Y_pred))
```

```python
#Comparing real values vs predicted values
df_compare = pd.DataFrame({'Text':X_test, 'Actual_Label':Y_test, 'Predicted_Label':Y_pred})

df_compare = df_compare.loc[df_compare['Actual_Label'] != df_compare['Predicted_Label']]
print('Data Shape: ', df_compare.shape)
df_compare
```
