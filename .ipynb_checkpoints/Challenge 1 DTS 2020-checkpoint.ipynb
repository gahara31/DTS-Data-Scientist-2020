# a) How many calories does the average McDonald's value meal contain?
The key in this question is <b><i>meal</i></b>. It asked only for meal type menu. We then filter menu based on their category and find its average calories.  

```python
df_meal = df[(df['Category'] == 'Chicken & Fish') | (df['Category'] == 'Beef & Pork' ) | (df['Category'] == 'Breakfast' )]
print('Calories average contain on McDonald\'s value meal:', df_meal["Calories"].mean())
```

# b) How much do beverages, like soda or coffee, contribute to the overall caloric intake?
The key in this question is <b><i>Overal caloric intake</i></b>. We saw that there's no column for overall caloric intake, only amount of calories for each menu. So we had to do some research and found that a person needs 2000 calories a day (the value might be different based on age, gender,and daily activities, to simplify calculation we just use 2000). We created a new column that contains percentage of calories of each menu based on 2000 calories requirements then we filter the the dataframe to show only beverages types of menu. 

```python
#create new column for % calories intake 
#daily calories intake is 2000 calories
df['Calories (% Daily Value)'] = (df['Calories']/2000)* 100
```

```python
#filter dataframe
df_beverage = df.query('Category == "Coffee & Tea" | Category == "Smoothies & Shakes" | Category == "Beverages"') 
df_beverage = df_beverage.loc[:,['Category','Item','Calories','Calories (% Daily Value)']]
df_beverage.sort_values('Calories (% Daily Value)',ascending=False,inplace=True)
df_beverage

# #plot result
# x=df_beverage['Item']
# y=df_beverage['Calories (% Daily Value)']
# fig, ax = plt.subplots()
# ax.barh(x, y)
```

In the table above shown 5 top most beverages and 5 lower most beverage sorted by their contribution to daily caloric requirement

# c) Does ordered grilled chicken instead of crispy increase a sandwich's nutritional value?
The key in this question is <i><b>nutritional value</b></i> but there's no column explicitly shown it. After researching we come to conclusion that each column represent a part of nutritional value. We can represent nutritional value by summing up each column (beside columns that represent percentage (%) beause it use different scale). Then We filter for only menu that has "sandwich" in it and present it side by side to compare it easily. In case there's a menu that have no pair, we drop it. 

```python
#check which value is higher
def HigherNutValue (row):
   if row['Total Nutritional Grilled'] > row['Total Nutritional Crispy'] :
      return 'GRILLED'
   else :
      return 'CRISPY'

#select relevant data
df_sandwich = df[df['Item'].str.contains("Sandwich")]

#find total nutritional value
df_buffer= df_sandwich.loc[:,['Item','Calories','Total Fat','Trans Fat','Sodium','Carbohydrates','Dietary Fiber','Sugars','Protein']]
df_buffer['Total Nutritional'] = df_buffer.sum(axis=1)
df_buffer

#remove unwanted data
df_buffer.drop(labels=65, axis=0, inplace=True) 
df_buffer.drop(labels=['Calories','Total Fat','Trans Fat','Sodium','Carbohydrates','Dietary Fiber','Sugars','Protein'],axis=1,inplace=True)

#split 
df_grilled = df_buffer[df_buffer['Item'].str.contains("Grilled Chicken")]
df_crispy = df_buffer[df_buffer['Item'].str.contains("Crispy Chicken")]

#rename splitted & reset index
df_grilled.columns=('Item Grilled', 'Total Nutritional Grilled')
df_grilled.reset_index(drop=True, inplace=True)

df_crispy.columns=('Item Crispy', 'Total Nutritional Crispy')
df_crispy.reset_index(drop=True, inplace=True)

#combine table
df_compare = pd.concat([df_grilled, df_crispy],axis=1, join='inner')
df_compare['HigherNutValue'] = df_compare.apply (lambda row: HigherNutValue(row), axis=1)
df_compare.head()
```

The table above shows sandwich menu that contain Crispy Chicken and Grilled Chicked side by side along with its nutritional value. We also made some little function that automatically compare both and return which one has higher nutritional value. From the four pair menu compared, all sandwich that use Crispy chicken give more nutritional value

# d) What about ordering egg whites instead of whole eggs?
This question is pretty similar with the previous question, hence our approach is the same.

```python
#check which value is higher
def HigherNutValue (row):
   if row['Total Nutritional Egg White'] > row['Total Nutritional Egg Whole'] :
      return 'Egg White'
   else :
      return 'Egg Whole'

df_egg = df[df['Item'].str.contains("Egg")]

# find total nutritional value
df_buffer2= df_egg.loc[:,['Item','Calories','Total Fat','Trans Fat','Sodium','Carbohydrates','Dietary Fiber','Sugars','Protein']]
df_buffer2['Total Nutritional'] = df_buffer2.sum(axis=1)
df_buffer2.reset_index(drop=True,inplace=True)

#remove unwanted data
df_buffer2.drop(df_buffer2.index[[4,13,20,21,22,23,24]], axis=0, inplace=True) 
df_buffer2.drop(labels=['Calories','Total Fat','Trans Fat','Sodium','Carbohydrates','Dietary Fiber','Sugars','Protein'],axis=1,inplace=True)

#split 
df_eggWhite = df_buffer2[df_buffer2['Item'].str.contains("Egg White")]
df_eggWhole = df_buffer2[~df_buffer2['Item'].str.contains("Egg White")]

#rename splitted & reset index
df_eggWhite.columns=('Item Egg White', 'Total Nutritional Egg White')
df_eggWhite.reset_index(drop=True, inplace=True)

df_eggWhole.columns=('Item Egg Whole', 'Total Nutritional Egg Whole')
df_eggWhole.reset_index(drop=True, inplace=True)

#combine table
df_compare = pd.concat([df_eggWhite, df_eggWhole],axis=1)
df_compare['HigherNutValue'] = df_compare.apply (lambda row: HigherNutValue(row), axis=1)
df_compare.head(15)
```

The table above shows menu that contain White Egg and Whole Egg side by side along with its nutritional value. From the eight pair menu compared, all menu that use Whole Egg give more nutritional value

# e) What is the least number of items could you order from the menu to meet one day's nutritional requirements?
the key in this question is <i><b>least number of items</b></i> to fulfill daily nutritional requirements. For this question we use only column that is in percetage. Reason being is the percetage is already based on daily value requirements. Next we take the the top menu from each percentage column that already sorted. We then sum up each column to see whether all column have value >= 100% which means nutritional requirements for that column already fulfilled. 

```python
#remove unwanted column
df_buffer3= df.loc[:,['Item','Total Fat (% Daily Value)','Sodium (% Daily Value)','Carbohydrates (% Daily Value)','Dietary Fiber (% Daily Value)','Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)','Calcium (% Daily Value)', 'Iron (% Daily Value)','Calories (% Daily Value)']]

#create empty dataframe
df_buffer4 = df_buffer3
df_buffer4 = pd.DataFrame(columns=df_buffer4.columns)

# add the top menu from each column
df_buffer3.sort_values('Total Fat (% Daily Value)', ascending=False,inplace=True)
df_buffer4 = df_buffer4.append(df_buffer3.head(1))

df_buffer3.sort_values('Sodium (% Daily Value)', ascending=False,inplace=True)
df_buffer4 = df_buffer4.append(df_buffer3.head(1))

df_buffer3.sort_values('Carbohydrates (% Daily Value)', ascending=False,inplace=True)
df_buffer4 = df_buffer4.append(df_buffer3.head(1))

df_buffer3.sort_values('Dietary Fiber (% Daily Value)', ascending=False,inplace=True)
df_buffer4 = df_buffer4.append(df_buffer3.head(1))

df_buffer3.sort_values('Vitamin A (% Daily Value)', ascending=False,inplace=True)
df_buffer4 = df_buffer4.append(df_buffer3.head(1))

df_buffer3.sort_values('Vitamin C (% Daily Value)', ascending=False,inplace=True)
df_buffer4 = df_buffer4.append(df_buffer3.head(1))

df_buffer3.sort_values('Calcium (% Daily Value)', ascending=False,inplace=True)
df_buffer4 = df_buffer4.append(df_buffer3.head(1))

df_buffer3.sort_values('Iron (% Daily Value)', ascending=False,inplace=True)
df_buffer4 = df_buffer4.append(df_buffer3.head(1))

df_buffer3.sort_values('Calories (% Daily Value)', ascending=False,inplace=True)
df_buffer4 = df_buffer4.append(df_buffer3.head(1))

#drop duplicate data
df_buffer4.drop_duplicates(subset=['Item'],inplace=True)
df_buffer4

# check manually if sum of each column is >100
df_buffer5 = df_buffer4.sum(axis=0)
print(df_buffer5)
```

Above shown the sum of each percentage column for the 6 menu that give the most value for one or more column. We still lack dietary fiber by 5%. We need at least 1 more menu to fulfill it. So our conclusion is: it takes at least 7 menu to fulfill daily nutritional requirements
